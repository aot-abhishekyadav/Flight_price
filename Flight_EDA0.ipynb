```python
# spliting Date of journey into Date, Month and year
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
# Converting Date,Month, year from object to integer
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis = 1, inplace = True)
```

```python
df['Arrival_Time'].str.split(' ')
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
df.head(1)
```

```python
# Converting hour, min from object to integer
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
# Dropping arrival time
df.drop('Arrival_Time',axis = 1, inplace = True)
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)
```

```python
df.drop('Dep_Time', axis = 1, inplace = True)
```

```python
# converting Duration into hour
df['Duration_hr'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.head()
```

```python
df.drop(6474, axis = 0, inplace = True) # Dropping two row which have value of 5m in Duration axis = 0 mean delete row
df.drop(2660, axis = 0, inplace = True) # 6474,2660 are the row that contain sm value and we are deleting this so we can convert them in int type
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Airline'] = le.fit_transform(df['Airline'])
df['Source'] = le.fit_transform(df['Source'])
df['Destination'] = le.fit_transform(df['Destination'])
df['Additional_Info'] = le.fit_transform(df['Additional_Info'])
```

```python
df.head()
```

```python
# Drop rows where 'Price' (the target variable) is NaN
df.dropna(subset=['Price'], inplace=True)
# Now we have to test model
```

```python
X = df.drop('Price', axis = 1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

 # Using Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train, y_train)
```

```python
y_pred_rf = rf_regressor.predict(X_test)
r2_score(y_test, y_pred_rf)
```
